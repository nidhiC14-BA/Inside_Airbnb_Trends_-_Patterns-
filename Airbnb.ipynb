```python
df = df.dropna(subset=['NAME'])
```

```python
df["country"].fillna("United States", inplace=True)
```

```python
df["country code"].fillna("US", inplace=True)
```

```python
df["instant_bookable"].fillna("Not mentioned", inplace=True)
```

```python
df["cancellation_policy"].fillna("unknown", inplace=True)
```

```python
df["Construction year"].fillna("unknown", inplace=True)
```

```python
df["price"] = df["price"].replace('[₹$,]', '', regex=True)              
 
```

```python
df['price'] = df['price'].astype(float)
```

```python
df["price"].fillna(df["price"].median(), inplace=True)
```

```python
df['service fee' ] = df['service fee'].replace('[₹$,€]', '', regex=True)
```

```python
df['service fee'] = df['service fee'].astype(float)
```

```python
df["service fee"].fillna(df["service fee"].median(), inplace=True)
```

```python
df["minimum nights"].fillna(1, inplace=True)
```

```python
df["number of reviews"].fillna(0, inplace=True)
```

```python
df["reviews per month"].fillna(0, inplace=True)
```

```python
df["house_rules"].fillna("No house rules mentioned", inplace=True)
```

```python
df["license"].fillna("Not Provided", inplace=True)
```

```python
df.rename(columns={
    'host name': 'host_name',
    'neighbourhood group': 'neighbourhood_group',
    'country code': 'country_code',
    'room type':  'room_type',
    'Construction year':'Construction_year',
    'service fee':'service_fee',
    'minimum nights':'minimum_nights',
    'number of reviews':'number_of_review',
    'last review':'last_review',
    'reviews per month':'reviews_per_month',
    'review rate number':'review_rate_number',
    'calculated host listings count':'calculated_host_listings_count',
    'availability 365':'availability_365',
}, inplace=True)
```

```python
df['neighbourhood_group'] = df['neighbourhood_group'].replace({ 'manhatan': 'Manhattan','brookln': 'Brooklyn',})
```
